多维正态分布
- 观察 cov中 方差大小，相关系数方向。对3D、等高线的影响。

```python
from scipy.stats import multivariate_normal
import numpy as np
```

```python
mean = [0, 0]
cov = [
    [2, -0.5],
    [-0.5, 2]
]
```

```python
rv = multivariate_normal(mean= mean, cov=cov)
```

```python
x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)
x, y = np.meshgrid(x, y)
points = np.dstack((x, y))
probs = rv.pdf(points)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,4))
ax1 = fig.add_subplot(1, 2, 1, projection ='3d')
ax1.contour3D(x, y, probs, cmap = 'RdGy')
ax1.scatter3D(x, y, probs, s = 10)
ax1.set_title('3d')
ax2 =  fig.add_subplot(1, 2, 2)
ax2.contour(x, y, probs, cmap = 'binary')
ax2.set_title('contour prob')
fig.text(0.5, 0.01, f'mean = {mean}, cov = {cov}', ha='center', fontsize=12)
fig.tight_layout()
```
