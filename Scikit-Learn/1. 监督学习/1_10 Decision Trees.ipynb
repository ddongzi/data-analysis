# Decision Trees

- DTs are a **non-param supervised learning method** used for **classification** and **regression**.
- A set of IF-ELSE

### 1.10.1 Classification

```python
from sklearn.tree import DecisionTreeClassifier
```

Using iris dataset, construct a tree:

```python
from sklearn.datasets import load_iris
iris = load_iris()
X, y = iris.data, iris.target
clf = DecisionTreeClassifier()
clf.fit(X,y)
```

We can plot this tree

```python
X[0]
```

```python
from sklearn import tree
_ = tree.plot_tree(clf)
```
